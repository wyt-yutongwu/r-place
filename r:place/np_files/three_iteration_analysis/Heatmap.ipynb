```python
import matplotlib.pyplot as plt
import sys
sys.path.append("../Python_code") # go to parent dir
from canvas_vis import *
import matplotlib.colors
```

```python
#Beginning of the experiment
#begin_time_22 = 1648835050

#Extracting number of updates for all canvas
data = extract_canvas_num_updates('../data/sorted_tile_placements.csv', x_min=0, x_max=1000, y_min=0, y_max=1000)
```

```python
def heat_map_with_color_bar(canvas, output_file_name, min_val, max_val):

    '''
        Converts float numbers on canvas to RGB. max_val is assigned to red
        and min_val to blue.
    '''
    plt.figure(frameon=False, figsize = (100,100))
    fig, ax = plt.subplots()
    #heatmap = ax.pcolor(data, cmap=cMap)

    ax.axis('off')
#    cMap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["blue","green","red"])

    plt.imshow(data, cmap = 'viridis',vmin = min_val, vmax = max_val)
    cbar = plt.colorbar(pad = 0.01, label = '# updates')

    cbar.set_ticks([min_val,(max_val + min_val) / 2,max_val])
    cbar.set_ticklabels([int(min_val), int((max_val + min_val)/2), ">" + str(int(max_val))])
    #High resolution is important
    plt.savefig(output_file_name, dpi=400, bbox_inches='tight')
```

```python
#We can't see much if we don't decrease the max color (red)
min_val = np.min(data)
max_val = np.mean(data) + np.std(data)

#Drawing activity on the canvas
heat_map_with_color_bar(data, "../plots/canvas_count_17.svg", min_val, max_val)
```
